4)Use the dictionary 'n_users' to plot a pie chart as shown below.

image.png

```python
# run this cell before proceeding further
n_users = {'Windows':323,'Android':403,'iOS':165,'OS X':71,'Linux':9}
```

```python
labels = []
sizes = []

for x, y in n_users.items():
    labels.append(x)
    sizes.append(y)

colors = ['#42b6f5','#95f542','#808080','#c0c0c0','#ffa500']    #got these color codes from the above given picture
 
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.0f%%')

plt.axis('equal')
plt.show()
```

5)The array 'over' contains the number of over and the array 'runs' contains the runs scored in each respective over.
Plot the total score for each over similar to the one shown below.

Plot.png

```python
# run this cell before proceeding further
over = np.arange(1,21)
over
```

```python
# run this cell before proceeding further
runs = np.random.randint(1,30,20)
runs
```

```python
cum_list = []   #creating a cumulative list
length = len(runs)   
cum_list = [sum(runs[0:x:1]) for x in range(0, length+1)]   
cum_list[1:]
scores=np.array(cum_list[1:])  #converting list to array
scores
```

```python
plt.scatter(over,scores)
plt.xticks(np.arange(0, len(over)+1, 1))
plt.plot(over,scores,c='orange')  
plt.show()
```

```python
# run this cell before proceeding further
import seaborn as sns
```

```python
import pandas as pd                                 #importing pandas
iris = pd.read_csv("iris.csv")                      #reading the file iris.csv
sns.set(style="darkgrid", color_codes=True)         # setting darkgrid
import warnings
warnings.filterwarnings("ignore")                 #ignoring warnings
```

1) Plot the heatmap of the correlation matrix of the iris data using seaborn.
Center the colorbar at 0 and display the data value in each cell.

```python
corr = iris.corr()       #creating correlation
sns.heatmap(corr, xticklabels=corr.columns.values,yticklabels=corr.columns.values,vmin=-1, vmax=1,center=0,annot=True)  #ploting heatmap
plt.show()
```

2) Using the iris data draw a boxplot 
- horizontally for each numeric variable in the dataset.
- for SepalWidthCm	 grouped by Species.

```python
sns.boxplot(data=iris, orient="h")
plt.show()
```

```python
sns.boxplot(x='Species',y='SepalWidthCm',data=iris)
plt.show()
```

3)Plot multiple pairwise bivariate distributions of the above dataset by taking "Species" as the Variable  to map plot aspects to different colors.

```python
sns.pairplot(iris, hue ='Species')
plt.show()
```

4)Replicate the below plot using seaborn library.
